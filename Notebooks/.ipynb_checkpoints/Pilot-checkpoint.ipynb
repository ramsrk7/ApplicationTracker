```python
# Modules that we are going to use
import imaplib
import email
import os
import shutil
import tempfile
import html2text
```

```python
# Credentials to login
LOGIN_EMAIL = "ramsrk7@gmail.com"
LOGIN_PWD = ""
SMTP_SERVER = "imap.gmail.com"
SMTP_PORT = 993
FROM_EMAIL = ''
```

```python
# Login to IMAP server and select folder from which we have to read
m = imaplib.IMAP4_SSL(SMTP_SERVER, SMTP_PORT)
m.login(LOGIN_EMAIL, LOGIN_PWD)
```

```python
m.select('"inbox"')
```

```python
mail_ids = []

status, data = m.search(None, 'ALL')
for block in data:
    mail_ids += block.split()
```

```python
j = 0
for i in mail_ids:
    status, data = m.fetch(i, '(RFC822)')
    j+=1
    
    if j==10:
        break
```

```python
mail_ids = mail_ids[::-1]
```

```python
j = 0
for i in mail_ids:
    status, data = m.fetch(i, '(RFC822)')
    j+=1
   

    for response_part in data:
        #print(response_part)
        # so if its a tuple...
        if isinstance(response_part, tuple):
            # we go for the content at its second element
            # skipping the header at the first and the closing
            # at the third
            message = email.message_from_bytes(response_part[1])

            # with the content we can extract the info about
            # who sent the message and its subject
            mail_from = message['from']
            mail_subject = message['subject']

            # then for the text we have a little more work to do
            # because it can be in plain text or multipart
            # if its not plain text we need to separate the message
            # from its annexes to get the text
            #print(message.is_multipart())
            if message.is_multipart():
                mail_content = ''

                # on multipart we have the text message and
                # another things like annex, and html version
                # of the message, in that case we loop through
                # the email payload
                for part in message.get_payload():
                    # if the content type is text/plain
                    # we extract it
                    print(part.get_content_type())
                    if part.get_content_type() == 'text/plain' or part.get_content_type() == 'text/html' or part.get_content_type() == 'multipart/related':
                        try:
                            print("try")
                            mail_content += part.get_payload()
                        except:
                            print("part", part.get_payload())
                            for subpart in part:
                                print(subpart)
                                mail_content += subpart
                                
            else:
                print("oops")
                # if the message isn't multipart, just extract it
                mail_content = message.get_payload()
                #print("Response part",response_part)

            # and then let's show its result
            print(f'From: {mail_from}')
            print(f'Subject: {mail_subject}')
            print(i)
            print(f'Content: {html2text.html2text(mail_content)}')
            print("##############")
            
    if j == 16:
        break

```
