### Emails regarding jobs sometimes comes without body but html. 

```python
def html2text(html, enc=0):
    
    if html == None:
        return None
    
    encodings = ['utf-8','utf-16','utf-32']
    
    try:
        html = html.decode(encodings[enc])
    except:
        if enc>2:
            return "Unable to Parse."
        html2text(html, enc+1)
        
    html = BeautifulSoup(html, 'html.parser')
    
    return html.get_text(separator=' ')
    
    
def preprocess_text(text):
    
    text = text.replace("\n","")
    text = text.replace("\r","")
    text = text.replace("\r\n","")
    text = text.replace("\n\r","")
    text = text.replace("\\r\\n","")
    text = text.replace("b'","")
    text = " ".join(text.split())
    
    return text
```

```python
workday = df[df['email'].str.contains('workday')].reset_index()
```

```python
workday['html'] = workday['html'].map(html2text)
```

```python
workday['html'] = workday['html'].map(preprocess_text)
```

```python
greenhouse = df[df['email'].str.contains('greenhouse-mail')].reset_index()
greenhouse['html'] = greenhouse['html'].map(html2text)
greenhouse['html'] = greenhouse['html'].map(preprocess_text)
greenhouse['body'].isna().value_counts()
```
